### Pandas series
A Pandas Series is like a column in a table.
It is a one-dimensional array holding data of any type. The axis labels are collectively called indexes.

```python
# Creating empty series
ser = pd.Series()

print(ser)
```

```python
# simple array
data = np.array(['g', 'e', 'e', 'k', 's'])
# print(data)
ser = pd.Series(data)
print(ser)
```

```python
# List
a = [9,1, 7, 2]
print(a)
type(a)
```

```python
# NP array
arr_a = np.array(a)
print(arr_a)
type(arr_a)
```

```python
# create a simple pandas series from a list
myvar = pd.Series(a)

print(myvar)
type(myvar)
```

#### Labels
If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.

#### Key/Value objects as series
You can also use a key/value object, like a dictionary, when creating a Series.

```python
# Create a simple Pandas Series from a dictionary
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
type(myvar)
```

To select only some of the items in the dictionary, use the index argument and specify only the items you want to include in the Series.

```python
calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"])

print(myvar)
```

Named Indexes

 With the index argument, you can name your own indexes.

```python
# Add a list of names to give each row a name
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df2 = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(df2)
```

## Matplotlib
To make necessary statistical inferences, it becomes essential to visualize your data, and Matplotlib is one such solution for Python users. It is a comprehensive plotting library useful for those working with Python and NumPy.
- Matplotlib is an open-source drawing library that supports various drawing types
- You can generate plots, histograms, bar charts, and other types of charts with just a few lines of code
- It’s often used in web application servers, shells, and Python scripts

Pyplot is a Matplotlib module that provides simple functions for adding plot elements, such as lines, images, text, etc. to the axes in the current figure.

```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[1,2,3]
# plotting our canvas
plt.plot(x,y )
plt.title('X vs Y')
# showing our plot
plt.show()

```

With the subplot() function you can draw multiple plots in one figure

```python
# Draw 2 plots:

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.title('A')
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)
plt.title('B')
plt.show()
```

```python
# Draw 2 plots on top of each other
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)
plt.title('graphs on top of each other')
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
# Add a title for the entire figure:

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

Histogram

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.
hist() is used to plot histograms

```python
# A simple histogram:
x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

Creating Pie Charts

With Pyplot, you can use the pie() function to draw pie charts

```python
# A simple pie chart:
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

```python
# A labelled pie chart:
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
# Add a legend with a header
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```

## Seaborn
It is also a Python library used for plotting graphs with the help of Matplotlib, Pandas, and Numpy. It is built on the roof of Matplotlib and is considered as a superset of the Matplotlib library.

```python
import seaborn as sns
import warnings
# Plotting a distplot
sns.distplot([0, 1, 2, 3, 4, 5])

plt.show()
```

```python
# Plotting a Distplot Without the Histogram

sns.distplot([0, 1, 2, 3, 4, 5], hist=False)

plt.show()
```
